```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('acc_players-2324F.csv', header=1)
df.head(5)
```

```python
print(df.PTS.sum())
```

```python
df[df.MP == df.MP.max()]
```

```python
df.sort_values(by='TRB', ascending=False).head(5)
```

```python
df_500_min = df[df.MP > 500]
df_500_min
```

```python
df[df.AST == df.AST.max()]
```

```python
df.sort_values(by='AST', ascending=False).head(3)
```

```python
df.sort_values(by='BLK', ascending=False).head(3)
```

```python
df_school = df.groupby(['School'])
df_school.PTS.sum()
```

```python
df_school.AST.sum()
```

```python
df_school.sum().sort_values(by='PTS', ascending=False).PTS.head(3)
```

```python
top_5_players = df.sort_values(by='PTS', ascending=False).head(5)

sns.set(style='whitegrid')
sns.barplot(x='Player', y='PTS', data=top_5_players)
plt.title('Top 5 Players by Total Points')
plt.show()
```

```python
sns.set(style='whitegrid')
sns.regplot(x='FG%', y='WS', data=df)
plt.title('Field Goal Percentage vs. Win Shares')
plt.show()
```

```python
df[['FG%', 'WS']].corr()
```
